```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
confirmed = pd.DataFrame({
    'Country/Region':['India','US','Brazil','Russia','UK'],
    '1/1/21':[10000,50000,30000,20000,15000],
    '1/2/21':[10500,53000,32000,21000,16000],
    '1/3/21':[11000,56000,34000,22000,17000],
    '1/4/21':[11800,59000,36000,23000,18000],
    '1/5/21':[12500,62000,38000,24000,19000]
})

deaths = pd.DataFrame({
    'Country/Region':['India','US','Brazil','Russia','UK'],
    '1/1/21':[200,1000,700,400,300],
    '1/2/21':[220,1100,750,420,320],
    '1/3/21':[240,1200,800,440,340],
    '1/4/21':[260,1300,850,460,360],
    '1/5/21':[280,1400,900,480,380]
})

recovered = pd.DataFrame({
    'Country/Region':['India','US','Brazil','Russia','UK'],
    '1/1/21':[8000,40000,25000,15000,12000],
    '1/2/21':[8500,42000,27000,16000,13000],
    '1/3/21':[9000,44000,29000,17000,14000],
    '1/4/21':[9500,46000,31000,18000,15000],
    '1/5/21':[10000,48000,33000,19000,16000]
})
```

```python
confirmed
```

```python
latest_date = '1/5/21'

summary = pd.DataFrame({
    'Confirmed': confirmed[latest_date],
    'Deaths': deaths[latest_date],
    'Recovered': recovered[latest_date]
}, index=confirmed['Country/Region'])

summary
```

```python
summary['Death Rate (%)'] = (
    summary['Deaths'] /
    summary['Confirmed']
) * 100

summary['Recovery Rate (%)'] = (
    summary['Recovered'] /
    summary['Confirmed']
) * 100

summary
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=summary['Confirmed'],
    y=summary.index
)

plt.title("Confirmed Cases by Country")
plt.show()
```

```python
india = [10000,10500,11000,11800,12500]

dates = [
    '1/1/21',
    '1/2/21',
    '1/3/21',
    '1/4/21',
    '1/5/21'
]

plt.figure(figsize=(8,5))
plt.plot(dates, india, marker='o')

plt.title("COVID Cases Trend in India")
plt.xlabel("Date")
plt.ylabel("Cases")

plt.show()
```

```python
countries = ['India','US','Brazil','Russia','UK']

cases = summary['Confirmed']

plt.figure(figsize=(8,5))
plt.bar(countries, cases)

plt.title("Country Comparison")
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    summary[
        ['Confirmed','Deaths','Recovered']
    ].corr(),
    annot=True
)

plt.title("Correlation Heatmap")
plt.show()
```
